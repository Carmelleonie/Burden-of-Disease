```python
from Modules.prepare_data import *
from Modules.death_case_for_country import *
from Modules.death_cases_in_year import *
from Modules.country_and_year import *
from dash import html, dcc, Output, Input, callback
from jupyter_dash import JupyterDash
```

```python
# Get Data
filename = "data/cause_of_deaths.csv"
df = prepare_data(filename=filename)
df.head()
```

```python
country_name = list(df["Country/Territory"].value_counts().index)
country_name
```

```python
# Create an app
app = JupyterDash(__name__)
```

```python
# Create the app's layout
app.layout = html.Div(
    [
        html.H1(children="Global Mortality Insights: Historical Dataset on Causes of Death",
               style={
                   "textAlign": "justify",
                   "color": "blue"
               }),
        html.Div([
            html.H2(children="Trend of death over time (1990–2019)"),
            dcc.Dropdown(id="country-name", options=[{"label":country, "value":country}for country in country_name],
                         value="Benin"),
            dcc.Graph(id="trend-of-country")
        ])
    ]
)

@callback(Output(component_id="trend-of-country", component_property="figure"),
         Input(component_id="country-name", component_property="value"))
def death_trend_in_country(country):
    year, top_causes = prepare_data_country(df, coutry_name=country)
    fig = country_death_trend(year)
    return fig
```

```python
# Run the app
if __name__ == "__main__":
    app.run(mode="external", debug=True)
```
